```python
from docx import Document
import re
import pandas as pd
import os

# 1. Đọc file Word
path = "/content/Luật Lao Động.docx"   # đổi tên file cho đúng
doc = Document(path)

# Lấy toàn bộ paragraph nhưng loại paragraph rỗng
paragraphs = [p.text.strip() for p in doc.paragraphs if p.text.strip() != ""]

# 2. Ghép thành danh sách để duyệt tuần tự
records = []
current_title = None
current_content = []

for line in paragraphs:

    # Dòng tiêu đề Điều
    if re.match(r"^Điều\s+\d+", line):

        # Nếu đã có điều trước đó -> lưu lại
        if current_title:
            records.append([current_title, " ".join(current_content).strip()])

        # Reset cho điều mới
        current_title = line.strip()
        current_content = []

    # Dòng nội dung
    else:
        current_content.append(line.strip())

# Thêm điều cuối cùng
if current_title:
    records.append([current_title, " ".join(current_content).strip()])

# 3. Xuất file Excel
df = pd.DataFrame(records, columns=["Điều", "Nội dung"])
df["Bộ luật"] = "Dân sự" #Đổi này tương ứng với bộ luật mà mình đang muốn crawl

df.to_excel("Bo_luat_Lao_Dong_final.xlsx", index=False)

print("Hoàn tất! Đã tạo file Bo_luat_Lao_Dong_final.xlsx")
```
